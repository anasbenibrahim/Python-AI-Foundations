# **Converting integer to float**


```python
integer_number = 12345
float_number = 1.2345
float_number2 = 3.3

new_number = integer_number + float_number
new_number2 = new_number + float_number2

print("Value:",new_number)
print("Data Type:",type(new_number))
print("New Value:",new_number2)
```

# **Explicit Type Conversion**


```python
num_string = '12'
num_integer = 23

print("Data type before Type Casting:",type(num_string))

# explicit type conversion
num_string = int(num_string)

print("Data type before Type Casting:",type(num_string))

num_sum = num_integer + num_string

print("Sum:",num_sum)
print("Data type of sum:",type(num_sum))
```

# **Python Operators**

## ***Example 1: Arithmetic Operators in Python***


```python
a = 8
b = 4

# Addition
print ('Sum: ', a + b)

# Subtraction
print ('Subtraction: ', a - b)

# Multiplication
print ('Multiplication: ', a * b)

# Division
print ('Division: ', a / b)

# Floor Division
print ('Floor Division: ', a // b)

# Modulo
print ('Modulo: ', a % b)

# a To The Power b
print ('Power: ', a ** b)
```

## ***Example 2: Assignment Operators***

```python
# Assign 10 to a
a = 10

# Assign 5 to b
b = 5

# Assign the sum of a and b to a
a += b  # a = a + b

print(a)
```

## ***Example 3: Comparison Operators***

```python
a = 2
b = 3

# Equal to operator
print('a == b =', a == b)

# Not equal to operator
print('a != b =', a != b)

# Greater than operator
print('a > b =', a > b)

# Less than operator
print('a < b =', a < b)

# Greater than or equal to operator
print('a >= b =', a >= b)

# Less than or equal to operator
print('a <= b =', a <= b)
```

## ***Example 4: Logical Operators***

```python
# Logical AND
print(True and True)
print(True and False)

# Logical OR
print(True or False)

# Logical NOT
print(not True)
```

## ***Example 6: Python Special Operators***

```python
x1 = 5
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1,2,3]
y3 = [1,2,3]

print(x1 is not y1)

print(x2 is y2)

print(x3 is y3)
```

## ***Example 7: Membership Operators***

```python
message = 'Hello world'
dict1 = {1:'a', 2:'b'}

# Check if 'H' is present in message string
print('H' in message)

# Check if 'hello' is present in message string
print('hello' not in message)

# Check if '1' key is present in dict1
print(1 in dict1)

# Check if 'a' key is present in dict1
print('a' in dict1)
```
