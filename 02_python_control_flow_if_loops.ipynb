# **Python If...Else Statement**

## ***Example 1: Python if Statement***

```python
number = 10

# Check if number is greater than 0
if number > 0:
  print('Number is positive')

print('This statement always executes')
```

## ***Example 2: if...else Statement***

```python
number = -2

if number > 0:
  print('Positive number')

else:
  print('Negative number')

print('This statement always executes')
```

## ***Example 3: if...elif...else Statement***

```python
number = 0

if number > 0:
  print('Positive number')

elif number < 0:
  print('Negative number')

else:
  print('Zero')

print('This statement always executes')
```

```python
number = 5

# Outer if statement
if number >= 0:
  # inner if statement
  if number == 0:
    print('Number is 0')

  # inner else statement
  else:
    print('Number is positive')

# Outer else statement
else:
  print('Number is negative')
```

```python
number = 11

if number > 0:
  print('Positive')

number = 12

if number > 0: print('Positive')
```

```python
grade = 30

if grade >= 50:
  result = 'pass'
else:
  result = 'fail'

print(result)

grade = 50

result = 'pass' if grade >= 50 else 'fail'

print(result)
```

```python
age = 35
salary = 6000

# Add two conditions using and operator
if age >= 30 and salary >= 5000:
  print('Eligible for the premium membership')
else:
  print('Not eligible for the premium membership')
```

# **Python for Loop**

```python
languages = ['Swift', 'Python', 'Java']

# Access elements of the list one by one
for i in languages:
  print(i)
```

## ***Example 1: Loop Through A String***

```python
language = 'Python'

# Interate over each character in language
for x in language:
  print(x)
```

## ***Example 2: For Loop with Python range()***

```python
# Iterate from i = 0 to i = 3
for i in range(4):
  print(i)
```

## ***Example 3: Python for loop with else clause***

```python
digits = [0, 1, 5, 3, 4]

for i in digits:
  print(i)
else:
  print("No items left.")
```

# **Python while Loop**

```python
number = 1

while number <= 3:
  print(number)
  number = number + 1
```

## ***Example 1: Python while loop***

```python
# Calculate the sum of numbers until user enters 0
number = int(input('Enter a number: '))

total = 0

# Iterate until the user enters 0
while number != 0:
  total += number
  number = int(input('Enter a number: '))

print('The sum is', total)
```

## ***Example 2: Python while loop with break statement***

```python
while True:
  user_input = input('Enter your name: ')

  # Terminate the loop when user enters end
  if user_input == 'end':
    print(f'The loop is ended')
    break

  print(f'Hi {user_input}')
```

```python
counter = 0

while counter < 2:
  print('This is inside loop')
  counter = counter + 1
else:
  print('This is inside else block')
```

```python
while True:
  user_input = input("Enter password: ")

  # Terminate the loop when user enters exit
  if user_input == 'exit':
    print(f'Status: Entry Rejected')
    break

  print(f'Status: Entry Allowed')
```

# **Python continue Statement**

```python
for i in range(5):
  if i == 3:
    continue
  print(i)
```
